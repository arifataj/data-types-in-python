The union () methode returns a new set with all items from both sets

```python
names={'ajay','shinas','prenav','abishek'}
names1={'shinas','aswanth','pranav','abishek','ajay'}

names3=set.union(names,names1)
print(names3)
```

The different methode will retuns a new sets that will contain only the items from the sets s that present in the others set

```python
names={'ajay','shinas','prenav','abishek'}
names1={'shinas','aswanth','pranav','abishek','ajay'}

names3=set.intersection(names,names1)
print(names3)
```
